# Coleta de Dados — Capacidade Solar Instalada (ANEEL/SIGA)

Este notebook coleta os dados de capacidade instalada de energia solar
fotovoltaica no Brasil a partir do SIGA da ANEEL.

**Fonte:** Agência Nacional de Energia Elétrica (ANEEL) — SIGA  
**Acesso:** Portal de Dados Abertos da ANEEL  
**Granularidade:** Por empreendimento, UF, fase  
**Período:** Histórico completo

```python
import pandas as pd
import requests
from io import BytesIO
import warnings
warnings.filterwarnings('ignore')

print("Bibliotecas carregadas!")
```

```python
# URL do arquivo SIGA da ANEEL — atualizado mensalmente
url_siga = "https://dadosabertos.aneel.gov.br/dataset/siga-sistema-de-informacoes-de-geracao-da-aneel/resource/11ec447d-698d-4ab8-977f-b424d5deee6a"

# Buscar URL direta do CSV
r = requests.get(url_siga)
print(f"Status: {r.status_code}")
print(f"Tamanho: {len(r.content):,.0f} bytes")
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.content, 'html.parser')

# Procurar links de download
links = []
for a in soup.find_all('a', href=True):
    href = a['href']
    if any(ext in href.lower() for ext in ['.csv', '.xlsx', '.xls', '.zip']):
        links.append(href)
        print(href)

print(f"\nTotal de links encontrados: {len(links)}")
```

```python
# Baixar o CSV do SIGA
url_csv = "https://dadosabertos.aneel.gov.br/dataset/6d90b77c-c5f5-4d81-bdec-7bc619494bb9/resource/11ec447d-698d-4ab8-977f-b424d5deee6a/download/siga-empreendimentos-geracao.csv"

r_csv = requests.get(url_csv)
df_siga = pd.read_csv(BytesIO(r_csv.content), sep=';', encoding='latin1')

print(f"Shape: {df_siga.shape}")
print(f"\nColunas: {df_siga.columns.tolist()}")
print(f"\nPrimeiras linhas:")
df_siga.head()
```

```python
# Verificar tipos de fonte disponíveis
print("Tipos de geração:")
print(df_siga['SigTipoGeracao'].value_counts())

print("\nFontes de combustível:")
print(df_siga['NomFonteCombustivel'].value_counts().head(20))

print("\nFases das usinas:")
print(df_siga['DscFaseUsina'].value_counts())
```

```python
# Filtrar apenas solar fotovoltaica em operação
df_solar = df_siga[
    (df_siga['SigTipoGeracao'] == 'UFV') &
    (df_siga['DscFaseUsina'] == 'Operação')
].copy()

print(f"Usinas solares em operação: {len(df_solar):,.0f}")

# Converter data de entrada em operação
df_solar['DatEntradaOperacao'] = pd.to_datetime(
    df_solar['DatEntradaOperacao'], 
    format='%Y-%m-%d',
    errors='coerce'
)

# Verificar período disponível
print(f"\nPrimeira usina: {df_solar['DatEntradaOperacao'].min()}")
print(f"Última usina: {df_solar['DatEntradaOperacao'].max()}")

# Capacidade total instalada
capacidade_total = df_solar['MdaPotenciaFiscalizadaKw'].sum() / 1_000_000
print(f"\nCapacidade total instalada: {capacidade_total:.2f} GW")
```

```python
# Investigar datas inválidas
print("Distribuição de anos de entrada em operação:")
df_solar['ano_operacao'] = df_solar['DatEntradaOperacao'].dt.year
print(df_solar['ano_operacao'].value_counts().sort_index().head(20))

# Verificar registros com data inválida
invalidos = df_solar[df_solar['ano_operacao'] < 2000]
print(f"\nRegistros com data inválida (antes de 2000): {len(invalidos)}")
print(invalidos[['NomEmpreendimento', 'DatEntradaOperacao', 'MdaPotenciaFiscalizadaKw']].head())
```

```python
# Verificar outros datasets disponíveis na ANEEL
url_datasets = "https://dadosabertos.aneel.gov.br/dataset/"
r_datasets = requests.get(url_datasets)
soup_datasets = BeautifulSoup(r_datasets.content, 'html.parser')

# Procurar datasets relacionados a geração distribuída
datasets = []
for item in soup_datasets.find_all('h3', class_='dataset-heading'):
    titulo = item.text.strip()
    if any(termo in titulo.lower() for termo in ['distribu', 'micro', 'mini', 'solar', 'fotovolt']):
        datasets.append(titulo)

print("Datasets relacionados encontrados:")
for d in datasets:
    print(f"  {d}")
```

```python
# Baixar geração distribuída (telhados residenciais e comerciais)
url_gd = "https://dadosabertos.aneel.gov.br/dataset/5e0fafd2-21b9-4d5b-b622-40438d40aba2/resource/b1bd71e7-d0ad-4214-9053-cbd58e9564a7/download/empreendimento-geracao-distribuida.csv"

r_gd = requests.get(url_gd)
df_gd = pd.read_csv(BytesIO(r_gd.content), sep=';', encoding='latin1')

print(f"Shape: {df_gd.shape}")
print(f"\nColunas: {df_gd.columns.tolist()}")
print(f"\nPrimeiras linhas:")
df_gd.head()
```

```python
# Verificar tipos de geração e fontes
print("Tipos de geração:")
print(df_gd['SigTipoGeracao'].value_counts())

print("\nFontes de geração:")
print(df_gd['DscFonteGeracao'].value_counts())

print("\nPeríodo disponível:")
print(f"  Início: {df_gd['AnmPeriodoReferencia'].min()}")
print(f"  Fim: {df_gd['AnmPeriodoReferencia'].max()}")
```

```python
# Filtrar só solar
df_solar_gd = df_gd[df_gd['SigTipoGeracao'] == 'UFV'].copy()

# Verificar coluna de data de conexão
print("Amostra da coluna de atualização:")
print(df_solar_gd['DthAtualizaCadastralEmpreend'].head(10))

# Verificar se existe coluna de data de conexão
colunas_data = [c for c in df_solar_gd.columns if 'Dat' in c or 'Dth' in c or 'data' in c.lower()]
print(f"\nColunas de data disponíveis: {colunas_data}")
```

```python
# Converter data de conexão
df_solar_gd['data_conexao'] = pd.to_datetime(
    df_solar_gd['DthAtualizaCadastralEmpreend'], 
    errors='coerce'
)

# Criar colunas de ano e mês
df_solar_gd['ano'] = df_solar_gd['data_conexao'].dt.year
df_solar_gd['mes'] = df_solar_gd['data_conexao'].dt.month

# Filtrar período relevante (2021 em diante para alinhar com SENATRAN)
df_solar_gd = df_solar_gd[df_solar_gd['ano'] >= 2012].copy()

# Agrupar por mês — capacidade instalada acumulada
df_solar_mensal = df_solar_gd.groupby(['ano', 'mes']).agg(
    capacidade_kw=('MdaPotenciaInstaladaKW', 'sum'),
    instalacoes=('CodEmpreendimento', 'count')
).reset_index()

# Criar coluna de data
df_solar_mensal['data'] = pd.to_datetime(
    df_solar_mensal['ano'].astype(str) + '-' + 
    df_solar_mensal['mes'].astype(str) + '-01'
)

# Calcular capacidade acumulada em GW
df_solar_mensal = df_solar_mensal.sort_values('data').reset_index(drop=True)
df_solar_mensal['capacidade_gw_acumulada'] = df_solar_mensal['capacidade_kw'].cumsum() / 1_000_000

print(f"Período: {df_solar_mensal['data'].min()} até {df_solar_mensal['data'].max()}")
print(f"Total de instalações: {df_solar_mensal['instalacoes'].sum():,.0f}")
print(f"Capacidade total acumulada: {df_solar_mensal['capacidade_gw_acumulada'].iloc[-1]:.2f} GW")
print(f"\nÚltimos 12 meses:")
print(df_solar_mensal.tail(12)[['data', 'instalacoes', 'capacidade_gw_acumulada']])
```

```python
# Converter potência para numérico
df_solar_gd['MdaPotenciaInstaladaKW'] = pd.to_numeric(
    df_solar_gd['MdaPotenciaInstaladaKW'].astype(str).str.replace(',', '.'),
    errors='coerce'
)

print(f"Tipo da coluna: {df_solar_gd['MdaPotenciaInstaladaKW'].dtype}")
print(f"Valores nulos: {df_solar_gd['MdaPotenciaInstaladaKW'].isna().sum()}")
print(f"Amostra: {df_solar_gd['MdaPotenciaInstaladaKW'].head()}")
```

```python
# Salvar dados brutos e série temporal
df_solar_gd.to_csv('../data/raw/aneel_solar_gd_raw.csv', index=False)
df_solar_mensal.to_csv('../output/aneel_solar_serie_temporal.csv', index=False)

# Salvar também o SIGA centralizado
df_solar_filtrado = df_solar[df_solar['ano_operacao'] >= 2012].copy()
df_solar_filtrado.to_csv('../data/raw/aneel_siga_centralizado_raw.csv', index=False)

print("Arquivos salvos!")
print(f"\nResumo ANEEL:")
print(f"  Geração Distribuída:")
print(f"    Período: 2012 a Mar/2026")
print(f"    Instalações: {df_solar_mensal['instalacoes'].sum():,.0f}")
print(f"    Capacidade acumulada: {df_solar_mensal['capacidade_gw_acumulada'].iloc[-2]:.1f} GW")
print(f"\n  Geração Centralizada (SIGA):")
print(f"    Usinas em operação: {len(df_solar):,.0f}")
capacidade_centralizada = df_solar['MdaPotenciaFiscalizadaKw'].sum() / 1_000_000
print(f"    Capacidade: {capacidade_centralizada:.1f} GW")
print(f"\n  Total Brasil: ~{df_solar_mensal['capacidade_gw_acumulada'].iloc[-2] + capacidade_centralizada:.1f} GW")
```
